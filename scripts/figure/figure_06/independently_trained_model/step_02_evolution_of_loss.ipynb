```python
import re
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
trained_model_indicator_index = "multi_01"

notebook_path = os.getcwd() + "/"
loss_profile_path = notebook_path + "loss_evolution/"
panel_b_image_path = notebook_path + "panel_b/"
```

```python
directory = Path(loss_profile_path)
```

```python
epochs = []
train_geo_losses = []
valid_geo_losses = []
train_biCross = []
valid_biCross = []

for p in directory.glob("*.out"):
    file_string = str(p)
    with open(p, 'r') as file:
        for line_number, line in enumerate(file, start=1):
            # print(line.strip())  # .strip() removes newline characters
            if re.search(r'\btra_loss=\b', line, re.IGNORECASE):
                # print(f"Line {line_number}: Found 'train' -> {line.strip()}")
                result = re.split(r"[ep ;:;=;,;\n]", line)
                # print("train result", result)
                train_geo_losses.append( float(result[10]))
                train_biCross.append(float(result[15]))
                epochs.append(int(result[3]))            
                # print("train - geodesic loss:",  float(result[6]))
                
            if re.search(r'\bval_loss=\b', line, re.IGNORECASE):
                # print(f"Line {line_number}: Found 'valid' -> {line.strip()}")
                result = re.split(r"[=;,;\n]", line)
                # print("valid result", result)
                valid_geo_losses.append( float(result[3]))      
                valid_biCross.append(float(result[5]))
                # print("valid - geodesic loss:",  float(result[1]))

# Zip the lists together
combined = list(zip(epochs, train_geo_losses, valid_geo_losses))

# Sort by epoch (first element of each tuple)
combined.sort(key=lambda x: x[0])

# Unzip back into separate lists
epochs, train_geo_losses, valid_geo_losses = map(list, zip(*combined))
```

```python
np.max(train_geo_losses)
```

```python

fig, ax = plt.subplots(1, 1, figsize=(6.95, 6.6))
ax.plot(epochs, train_geo_losses, color = '#333333', linewidth = 1.1, label=r'$L_{geo}$ - train data')
ax.plot(epochs, valid_geo_losses, color = '#F52750', linewidth = 1.1, label=r'$L_{geo}$ - valid data')

ax.plot(epochs, train_biCross, color = '#2B2B2B', linestyle='--', linewidth = 1.4, label=r'$L_{cross}$ - train data')
ax.plot(epochs, valid_biCross, color = '#CC0E33', linestyle='--', linewidth = 1.4, label=r'$L_{cross}$ - valid data')

ax.set_xlabel(r"epoch", fontsize = 37)
ax.set_ylabel(r"geodesic loss, $L_{geo}$", fontsize = 37)
ax.legend(fontsize = 30)
ax.set_xlim([-1.5,255])
ax.set_ylim([-0.0000000001,0.63])
ax.set_yticks([0.0, 0.2, 0.4, 0.6])
formatter = ScalarFormatter(useOffset=False) # useOffset=False prevents an additional offset term
formatter.set_powerlimits((-2, -2)) # Force scientific notation for numbers within 10^-3 to 10^-3
formatter.set_scientific(True) # Ensure scientific notation is used
ax.yaxis.set_major_formatter(formatter)
ax.tick_params(axis='x', direction='out', length=12, width=0.8, color='k', labelsize=37)
ax.tick_params(axis='y', direction='out', length=12, width=0.8, color='k', labelsize=37)
plt.savefig(panel_b_image_path + trained_model_indicator_index + "_evolution_of_geodesic_loss_mulPred.pdf", bbox_inches='tight')
plt.show()
```
